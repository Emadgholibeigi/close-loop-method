```python
# library to make plot
import matplotlib.pyplot as plt
```

```python
# numpy as always
import numpy as np
```

```python
# library to use time
import time 
```

```python
# math library for cos , sin and tan2
import math 
```

```python
# initialize k_ru , k_alfa , K_beta ( Random )
k_ru   = 5
k_alfa = 32
k_beta = -0.1
```

```python
aaa =  k_alfa + ((5/3)*k_beta) - ((2/3.14) * k_ru)
aaa
```

```python
# initialize start robot positions x ,y in numpy array
#first column is x and second column is y
start_x_y = np.array([
    [ -50, 0 ],
    [-30 ,35 ],
    [ 40 , 35],
    [ 40 ,-35],
    [-35 ,-35]
])
```

```python
# Time step
dt = 0.001
```

```python
# a function to find zavieh robot
def Angel_check(x):
    if x>0 : 
        r = x - round(x/360) * 360
    else :
        r = x + round(abs(x)/360)*360
    return r    
```

```python
#to show position of robot and goal
def Show_robot( rx , ry , rt , gx , gy) :
    
    plt.scatter(gx , gy)
    
    return  plt.scatter( rx , ry ) 
```

```python
def move_to( from_loc_x , from_loc_y , to_loc_x , to_loc_y , teta ) :

    #robot start location initializing
    robot_x  = from_loc_x
    robot_y = from_loc_y
    
    #teta means robot angel 
    robot_teta = teta
    robot_teta = Angel_check(robot_teta)

    # initialize goal positions x ,y 
    goal_x = to_loc_x
    goal_y = to_loc_y
    
    # while loop condition instilize
    Flag = 1
    
    # plot lable 
    plt.xlabel('X')
    plt.ylabel('Y')

    #calculate loop
    while (Flag) :
        
        # show the current location of robot
        Show_robot(robot_x , robot_y , robot_teta , goal_x , goal_y)

        # calculate delta -> defference between goal and robot location
        delta_x = goal_x - robot_x
        delta_y = goal_y - robot_y

        # calculate ru , alfa and beta
        ru = math.sqrt( (math.pow(delta_x,2)) + (math.pow(delta_y,2)) ) 
        
        # loop end condition
        if ru < 0.01 :
            Flag = 0
            break
            
        alfa = -(robot_teta) + np.arctan2(delta_y,delta_x)
        alfa = Angel_check(alfa)

        beta = robot_teta - alfa
        beta = Angel_check(beta)
        
        # velocity
        v = k_ru * ru
        w = (k_alfa * alfa) + (k_beta * beta)

        # changes in ru , alfa and beta
        #--------------------------------------------
        ru_dot = -(k_ru * ru) * math.cos(alfa)
        alfa_dot = (k_ru * math.sin(alfa)) - (k_alfa * alfa) - (k_beta * beta)
        beta_dot = -(k_ru) * math.sin(alfa)

        if not( (-90 < alfa ) and ( alfa <= 90) ) :
            ru_dot = - (ru_dot)
            alfa_dot = - (alfa_dot)
            beta_dot = - (beta_dot)
        #--------------------------------------------    
        
        # update ru , alfa and beta
        ru = ru + (dt * ru_dot)
        alfa = alfa + (dt * alfa_dot)
        beta = beta + (dt * beta_dot)
        
        # velocity
        v= k_ru * ru
        w = (k_alfa * alfa) + (k_beta * beta)
        
        # update robot location
        robot_teta = robot_teta + (dt * w)
        robot_teta = Angel_check(robot_teta)

        robot_x = robot_x + (dt * v * math.cos(robot_teta))
        robot_y = robot_y + (dt * v * math.sin(robot_teta))
        
        print('ru = ' , ru)
        print('alfa = ' , alfa)
        print('beta = ' , beta)
        print('teta = ' , robot_teta)
        print('x = ' , robot_x)
        print('y = ' , robot_y)
        print('ru_dot = ' , ru_dot )
        
        time.sleep(0.001)
    for i in range(10) :
        print (' done *** please wait a moment ... ')
    
```

```python
for index in start_x_y :
    
    x = index[0]
    y = index[1]
    move_to ( x, y  , 0 , 0 , 0 )
```
